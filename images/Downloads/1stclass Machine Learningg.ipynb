## Visualization

```python
x = df['surface_covered_in_m2']
y = df['price_aprox_usd']

plt.xlabel("surface Area in Sq Meter")
plt.ylabel("Price in USD")

plt.scatter(x,y)
plt.show()
```

## Removing Discovered Outliers

```python
low, high = x.quantile([0.1, 0.9])
low, high
```

```python
df = df[df['surface_covered_in_m2'].between(low, high)]
df
```
